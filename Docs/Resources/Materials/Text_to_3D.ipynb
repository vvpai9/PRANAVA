```python
import torch
from meshgpt_pytorch import (
    MeshAutoencoder,
    MeshTransformer,
    mesh_render
)
from taylor_series_linear_attention import TaylorSeriesLinearAttn
import gradio as gr

def generate_mesh(input_text):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    transformer = MeshTransformer.from_pretrained("MarcusLoren/MeshGPT-preview").to(device)

    # Generate the mesh based on input text
    output = transformer.generate(texts=[input_text], temperature=0.0)

    # Save the rendering to an OBJ file
    file_path = './render.obj'
    mesh_render.save_rendering(file_path, output)

    return file_path

# Define the Gradio interface
def app():
    gr.Interface(
        fn=generate_mesh,
        inputs=gr.Textbox(label="Input Text", placeholder="Enter a description like 'bird'"),
        outputs=gr.Model3D(label="Generated 3D Mesh"),
        title="MeshGPT Renderer",
        description="Generate 3D mesh renderings based on text prompts using MeshGPT.",
    ).launch()

if __name__ == "__main__":
    app()
```
